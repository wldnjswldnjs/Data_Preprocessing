```python
import os # 운영체제 제어 모듈
import cv2 # opencv 모듈

dir = r'C:\Users\jiwon\OneDrive\바탕 화면\video'

fileEx = r'.mp4'

video_list = [os.path.join(dir, file) for file in os.listdir(dir) if file.endswith(fileEx)]

print(len(video_list)) # 동영상 파일 개수
```

```python
for video_file in video_list:
    video_capture = cv2.VideoCapture(video_file) # 동영상 파일 가져오기
    if not video_capture.isOpened(): # 파일이 열리지 않으면 메시지 출력
        print(f"Error opening video file: {video_file}")
        continue

    frame_count = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT)) # 동영상의 현재 프레임 수
    fps = int(video_capture.get(cv2.CAP_PROP_FPS)) # 프레임 속도 

    print(fps)

		# 초당 프레임 3 미만이면 해당 파일 종료 (이미지 추출 x)
    if fps < 3:
        print(f"Skipping video file: {video_file} (fps: {fps})")
        video_capture.release()
        continue

    # 이미지 파일명을 정의하기 위한 변수
    # basename 함수 : 파일 이름 추출
    # splitext 함수 : 파일 이름과 확장자 분리
    base_name = os.path.splitext(os.path.basename(video_file))[0]

    count = 0

    while video_capture.isOpened():
        ret, frame = video_capture.read()

        # ret : 프레임을 읽었는지 여부를 나타내는 불리언 값
        # frame : 실제 읽어온 이미지 프레임

        if not ret:
            break

        # 프레임당 이미지 추출
        if int(video_capture.get(1)) % (fps//3) == 0: # 초당 3장 추출
            print('Saved frame number : ' + str(int(video_capture.get(1))))
            file_name = base_name + '_frame%d.jpg' % count
            image_path = os.path.join(r'C:\Users\jiwon\OneDrive\바탕 화면\video_ex', file_name)  # 이미지 경로 생성
            os.makedirs(os.path.dirname(image_path), exist_ok=True)  # 디렉토리가 없으면 생성
            cv2.imwrite(image_path, frame)
            count += 1

    video_capture.release()

print("Image extraction completed.")
```
